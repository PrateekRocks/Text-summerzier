```python
punc=string.punctuation 
punc
```

```python
word_token= word_tokenize(text)
word_token
```

```python
print(word_token.count("transfusion"))
```

```python
for j in word_token:
    if j in punc:
        word_token.remove(j)
        
word_token
```

```python
for j in word_token:
    if j in stopword:
        word_token.remove(j)
        
word_token
```

```python
lemm = WordNetLemmatizer()
word_tokenise=[]
for i in word_token:
    i=i.lower()
    word_tokenise.append(lemm.lemmatize(i,pos="v"))
    
    
word_tokenise
```

```python
#frequency table
freqTable={}
for word in word_tokenise:
    if word in freqTable:
        freqTable[word] += 1
    else:
        freqTable[word] = 1
freqTable
```

```python
sentences_token=sent_tokenize(text)
sentences_token
```

```python
sentences_dict={}
for i in sentences_token:
    no_of_word=(len(word_tokenize(i)))
    for value in freqTable:
        if value in i.lower():
            if i in sentences_dict:
                sentences_dict[i] += freqTable[value]
            else:
                sentences_dict[i] = freqTable[value]
            
    sentences_dict[i]=sentences_dict[i]/no_of_word
    
sentences_dict
```

```python
sumvalues=0
for q in sentences_dict:
    sumvalues +=sentences_dict[q]
print(sumvalues)
average=int(sumvalues/ len(sentences_dict))
average
```

```python
large=sorted(sentences_dict.values() ,reverse=True)
```

```python
#summary
count=0
summary=""
for sent in sentences_token:
    if sent in sentences_dict and sentences_dict[sent] >average and count<5:
        summary +=" "+sent
        count +=1
summary
    
```
